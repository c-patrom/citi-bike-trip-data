```python
import pandas as pd
import os
```

#### Loading / Inspecting Citibike data

```python
df = pd.DataFrame()

for file in os.listdir('data'):
    if 'citibike' in file:
        file_path = os.path.join('data', file)
        temp_data = pd.read_csv(file_path)
        df = pd.concat([df, temp_data], ignore_index=True)

df.head()
```
